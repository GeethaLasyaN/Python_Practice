```python
#QUESTION 64 Write A Python Program To Find Scholarship For Students In Admission In A College On The Basis Of Student Marks.
Criteria To Follow
Marks Equal To 1100 Free Education
Marks > 1000 80% Monthly Fees Deduction Marks > 900 60% Monthly
Fees Deduction
Marks > 800 40% Monthly Fees Deduction
Marks > 700 30% Monthly Fees Deduction
Marks > 600 There Is No Scholarship


def find_scholarship(marks):
    if marks >= 1100:
        return "Free Education"
    elif marks > 1000:
        return "80% Monthly Fees Deduction"
    elif marks > 900:
        return "60% Monthly Fees Deduction"
    elif marks > 800:
        return "40% Monthly Fees Deduction"
    elif marks > 700:
        return "30% Monthly Fees Deduction"
    else:
        return "No Scholarship"

marks = int(input("Enter the student's marks: "))
scholarship = find_scholarship(marks)
print(f"The scholarship for the student is: {scholarship}")
```

```python
#QUESTION 70 Write A Python Program To Check The duplication of Student Identity In A Student List.
def check_duplication(student_list):
    unique_students = set(student_list)
    if len(unique_students) != len(student_list):
        return "Duplicate student identity found."
    else:
        return "No duplication in student identities."

students = []
for i in range(5):
    student_id = input(f"Enter student identity {i+1}: ")
    students.append(student_id)

result = check_duplication(students)
print(result)
```

```python
#QUESTION 96 Write A Python Program To Get 10 Data Item From User, How Much Integer, Float
And String Type Of Data Is Stored In A List
def categorize_data():
    integers, floats, strings = 0, 0, 0
    for i in range(10):
        data = input(f"Enter data item {i+1}: ")
        if data.isdigit():
            integers += 1
        elif data.replace('.', '', 1).isdigit():
            floats += 1
        else:
            strings += 1
    return integers, floats, strings

integers, floats, strings = categorize_data()
print(f"Integer count: {integers}, Float count: {floats}, String count: {strings}")
```

```python
#QUESTION 98 Write A Python Program To Count Prime Number From 1 To 100
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

prime_count = sum(1 for i in range(1, 101) if is_prime(i))
print(f"Number of prime numbers between 1 and 100: {prime_count}")

```
