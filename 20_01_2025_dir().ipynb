```python
#Right-angled triangle
rows=5
for i in range(1, rows+1):
  print("*"*i)
```

```python
#nverted right-angled triangle
rows=5
for i in range(rows, 0, -1):
  print("*"*i)
```

```python
#pyramid pattern
rows=5
for i in range(1,rows+1):
  print(" "*(rows-i)+"*"*(2*i-1))
```

```python
#Inverted Pyramid
rows = 5
for i in range(rows, 0, -1):
  print(" "*(rows-i)+"*"*(2*i-1))
```

```python
#rows=5
for i in range(1,rows+1):
  print(" "*(rows-i)+"*"*(2*i-1))
for i in range(rows, 0, -1):
  print(" "*(rows-i)+"*"*(2*i-1))

```

```python
#Rightangled triangle(Aligned to Right)
rows=5
for i in range(1, rows+1):
  print(" " *(rows-i)+"*"*i)
```

```python
#sandglass pattern
rows = 5
for i in range(rows, 0,-1):
  print(" "*(rows-i)+"*"*(2*i-1))
for i in range(1, rows+1):
  print(" "*(rows-i)+"*"*(2*i-1))
```

```python
#Hollow Pyramid
rows =5
for i in range(1, rows + 1):
  if i==1 or i==rows:
    print(" "*(rows-i)+"*"*(2*i-1))
  else:
    print(" " * (rows -i)+"*"+" "*(2*i-3)+"*")
```

```python
#Hollow dimond
rows = 5
for i in range(1, rows+1):
  if i == 1:
    print(" "*(rows-i)+"*")
  else:
    print(" "*(rows-i)+"*"+" "*(2*i-3)+"*")
for i in range(rows, 0, -1):
  if i==1:
    print(" "*(rows-i)+"*")
  else:
    print(" "*(rows-i)+"*"+" "*(2*i-3)+"*")
```

```python
#Number Pyramid
rows = 5
for i in range(1, rows + 1):
  print("  "*(rows-1)+"  ".join(str(j) for j in range(1, i+1)))
```
