```python
#COUNTING VOWELS IN A GIVEN WORD
vowel = ['a','e','i','o','u']
word = "programming"
count=0
for character in word:
  if character in vowel:
    count+=1
print(count)
```

```python
#Counting VOWELS in a Given Word
vowel = ['a','e','i','o','u']
word = "programming"
count=0
for character in word:
  if character not in vowel:
    count+=1
print(count)
```

```python
#COUNTING THE NUMBER OF OCCURANCES OF A CHARACTER IN A STRING
word = "programming"
character = 'g'
count = 0
for i in word:
    if i == character:
       count += 1
print(count)
```

```python
#COUNTING DIGITS,LETTERS,AND SPACES IN A STRING
import re
name = 'Python is 1'
digitCount = re.sub("[^0-9]","",name)
letterCount = re.sub("[^a-zA-Z]","",name)
spaceCount = re.sub("[^ ]","",name)
print(len(digitCount))
print(len(letterCount))
print(len(spaceCount))
```

```python
#COUNTING SPECIAL CHARACTERS IN A STRING
def count_sp_char(string):
    sp_char = "!@#$%^&*()|/:;<>'"
    count = 0
    for i in string:
        # The condition was changed.
        # You should check if the current charachter (i) exists within the sp_char string.
        if i in sp_char:
            count+=1
    return count
text = 'Hello!'
result = count_sp_char(text)
print(result)
```
